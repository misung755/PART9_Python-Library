# Python을 활용한 데이터분석(평가)

- 아래 코드셀을 실행해 **member** 데이터프레임을 만든 후 이후 코드셀을 완성하세요.

```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np

# 파일 읽어오기
file_path = 'https://raw.githubusercontent.com/Jangrae/csv/master/members.csv'
member = pd.read_csv(file_path, sep = ',')

# 상위 20개 행 확인
member.head(20)
```

**실습1]** NaN 값이 있는 행을 제거하세요.

```python
# NaN 값이 있는 행 제거
member.dropna(inplace=True)

# 확인
member.head(20)
```

**실습2]** 인덱스 값에 구멍이 없도록 인덱스를 초기화 하세요.

```python
# 인덱스 초기화
member.reset_index(drop=True, inplace=True)

# 확인
member.head(20)
```

**실습3]** Point 열과 Bonus 열을 더한 결과를 갖는 Result 라는 이름의 열을 추가하세요.

```python
# Result 열 추가
member['Result'] = member['Point'] + member['Bonus']

# 확인
member.head()
```

**실습4]** Grade 열에 대한 가변수(Dummy Variable)를 grade_dummy 라는 이름의 데이터프레임으로 만드세요. 

```python
# Dummy Variable 만들기
grade_dummy = pd.get_dummies(member['Grade'])

# 확인
grade_dummy.head()
```

**실습5]** grade_dummy 데이터프레임을 member 데이터프레임에 연결하세요.

```python
# 기존 데이터프레임에 붙이기
member = pd.concat([member, grade_dummy], axis=1, join='inner')

# 확인
member.head()
```

**실습6]** A --> Grade_A, B --> Grade_B, C --> Grade_C 로 열 이름을 변경하세요.

```python
# 열 이름 변경
member.rename(columns={'A': 'Grade_A',
                      'B': 'Grade_B',
                      'C':'Grade_C'}, inplace=True)

# 확인
member.head()
```

**실습7]** Age 열과 Grade 열을 제거하세요.

```python
# Age, Grade 열 제거
member.drop(['Age','Grade'], axis=1, inplace=True)

# 확인
member.head()
```

**실습8]** 전체 Result 평균에서 개인의 Result를 뺀 결괏값을 갖는 Diff_From_Mean 열을 Result 열 바로 뒤에 추가하세요.

```python
# Diff_From_Mean열 추가
Result_mean = member['Result'].mean() #6703.0
member.insert(4, 'Diff_From_Mean', Result_mean - member['Result'], allow_duplicates=True)

#확인
member.head(20)
```
